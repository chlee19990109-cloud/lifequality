```python
#최종 연소득(연속형)------------------------------------------------------------------------------------------

# 1. 조건부 월소득 77,99 제거
df = df[~df['cond_household_income_mth'].isin([77, 99])]

# 2. 이상코드 → 0
replace_zero = [77777,88888,99999]

df['household_income_yr'] = df['household_income_yr'].replace(replace_zero, 0)
df['household_income_mth'] = df['household_income_mth'].replace(replace_zero, 0)
df['cond_household_income_mth'] = df['cond_household_income_mth'].replace({88: 0})

# 3. 조건부 월소득 구간 → 각 구간의 중간값(만원)
cond_income_map = {
    1: 25, 2: 75, 3: 150, 4: 250,
    5: 350, 6: 450, 7: 550, 8: 650
}

df['cond_income_mth_value'] = df['cond_household_income_mth'].map(cond_income_map).fillna(0)

# 4. 최종 연소득 (만원 단위)
df['income_year_final'] = (
    df['household_income_yr']
    + df['household_income_mth'] * 12
    + df['cond_income_mth_value'] * 12
)

df['income_year_final'] = pd.to_numeric(
    df['income_year_final'],
    errors='coerce'
)

# 5.최종 연소득이 0인 경우 NAN 으로 만듬 -> 기술통계 구하기 위함
df.loc[df['income_year_final'] == 0, 'income_year_final'] = np.nan



#소득수준(범주형)-----------------------------------------------------------------------------------------------------------------
income_cat = pd.cut(
    df['income_year_final'],
    bins=[0, 2000, 4000, 6000, 8000, 10000, float('inf')],
    labels=['<2000', '2000–3999', '4000–5999', '6000–7999', '8000–9999', '≥10000']
)

df['income_cat'] = income_cat
```

#### 2. 건강행태(MET/수면/흡연/음주/칫솔질/체중조절)

```python
#MET-----------------------------------------------------------------------------------------------------------------
# 신체활동 관련 파생 변수 nan처리
na_codes = [77, 88, 99]
df['walk_mins'] = df['walk_mins'].replace(na_codes, np.nan)
df['walk_hours'] = df['walk_hours'].replace(na_codes, np.nan)
df['walk_days'] = df['walk_days'].replace(na_codes, np.nan)

df['mod_act_mins'] = df['mod_act_mins'].replace(na_codes, np.nan)
df['mod_act_hours'] = df['mod_act_hours'].replace(na_codes, np.nan)
df['mod_act_days'] = df['mod_act_days'].replace(na_codes, np.nan)

df['hig_act_mins'] = df['hig_act_mins'].replace(na_codes, np.nan)
df['hig_act_hours'] = df['hig_act_hours'].replace(na_codes, np.nan)
df['hig_act_days'] = df['hig_act_days'].replace(na_codes, np.nan)

# 걷기/중강도/고강도 총 시간(분)
def total_minutes(days, hours, mins):
    base = hours.fillna(0) * 60 + mins.fillna(0)
    return np.where(days.isna(), np.nan,
                    np.where(days == 0, 0, days * base))

df['walk_min_total'] = total_minutes(df['walk_days'], df['walk_hours'], df['walk_mins'])
df['mod_min_total']  = total_minutes(df['mod_act_days'], df['mod_act_hours'], df['mod_act_mins'])
df['hig_min_total']  = total_minutes(df['hig_act_days'], df['hig_act_hours'], df['hig_act_mins'])





# 걷기만 하는 부류
df['only_walk'] = np.where((df['walk_min_total'] >= 0) & (df['mod_min_total'] == 0) & (df['hig_min_total'] == 0), 0, 1)

# 총 MET점수
df['MET'] = np.where(
    df[['walk_min_total','mod_min_total','hig_min_total']].isna().any(axis=1),
    np.nan,
    df['walk_min_total']*3.3 + df['mod_min_total']*4.0 + df['hig_min_total']*8.0
)

# 신체활동 수준 구분(1=충분, 0=부족) : WHO 600 MET-min 기준
df['activity_level'] = np.where(
    df['MET'].isna(),
    np.nan,
    np.where(df['MET'] >= 600, 1, 0)
)
```

```python
#수면-----------------------------------------------------------------------------------------------------------------
# 수면 관련 파생변수
# 수면 시간(연속형)
df['sleep_hour'] = df['sleep_hour'].replace(na_codes, np.nan)
df['sleep_min']  = df['sleep_min'].replace(na_codes, np.nan)

df.loc[(df['sleep_hour'] < 0) | (df['sleep_hour'] > 23), 'sleep_hour'] = np.nan
df.loc[(df['sleep_min']  < 0) | (df['sleep_min']  > 59), 'sleep_min']  = np.nan

df['sleep_time'] = df['sleep_hour'] + df['sleep_min'] / 60
df['sleep_time'] = df['sleep_time'].round(2)

df.loc[(df['sleep_time'] < 3) | (df['sleep_time'] > 14), 'sleep_time'] = np.nan

# 수면 양호 여부(7~8시간이면 1, 그 외 0)
df['sleep_good'] = np.where(
    df['sleep_time'].isna(), np.nan,
    np.where(df['sleep_time'].between(7, 8, inclusive='both'), 1, 0)
)

# 수면 범주형
df['sleep_cat'] = pd.cut(
    df['sleep_time'],
    bins=[0, 6, 7, 8, 24],
    labels=['<6 h', '6–7 h', '7–8 h (normal)', '≥8 h'],
    right=False
)
```

```python
# PSQI 수면의 질

psqi_variables1 = ['edit_mtc_03z1','mtc_04z1','mtc_05z1','mtc_06z1', 'mtc_08z1','mtc_09z1','sleep_hour','sleep_min']
psqi_variables2 = ['mtc_12a1','mtc_12b1','mtc_12c1','mtc_12d1','mtc_12e1',
                       'mtc_12f1','mtc_12g1','mtc_12h1','mtc_12i1','mtc_12j1',
                       'sleep_qol','mtc_14z1','mtc_15z1','mtc_16z1']

# ======================================================
# 1. 유효 응답 여부 판단
#    variables1: 77, 99 중 하나라도 있으면 제외
#    variables2: 7, 9 중 하나라도 있으면 제외
# ======================================================

df['PSQI'] = (
    ~df[psqi_variables1].isin([77, 99]).any(axis=1)
    &
    ~df[psqi_variables2].isin([7, 9]).any(axis=1)
    )
# any(axis = 1) >> 값 두 종류 중 하나라도 있을시 제외

# ======================================================
# 2. 시간 변수 분(min) 단위 변환 (유효 응답만)
# ======================================================
df.loc[df['PSQI'], 'sleep_latency_min'] = (
    df['mtc_05z1'] * 60 + df['mtc_06z1']
) # 수면 소요시간

df.loc[df['PSQI'], 'sleep_time_min'] = (
    df['sleep_hour'] * 60 + df['sleep_min']
) # 실제 수면시간

df.loc[df['PSQI'], 'bed_time_min'] = (
    df['edit_mtc_03z1'] * 60 + df['mtc_04z1']
) # 잠자는 시각

df.loc[df['PSQI'], 'wake_time_min'] = (
    df['mtc_08z1'] * 60 + df['mtc_09z1']
) # 기상 시각

# 자정 이후 기상 보정
df.loc[
    (df['PSQI']) &
    (df['wake_time_min'] < df['bed_time_min']),
    'wake_time_min'
] += 1440

df.loc[df['PSQI'], 'time_in_bed_min'] = (
    df['wake_time_min'] - df['bed_time_min']
)

# ======================================================
# 3. C1 ~ C7 계산 (한국판 PSQI-K 기준)
# ======================================================

# C1. 주관적 수면의 질
df.loc[df['PSQI'], 'C1'] = df['sleep_qol'] - 1

# C2. 수면 잠복기
def latency_score(x):
    if x <= 15: return 0
    elif x <= 30: return 1
    elif x <= 60: return 2
    else: return 3

df.loc[df['PSQI'], 'latency_score1'] = (
    df.loc[df['PSQI'], 'sleep_latency_min']
    .apply(latency_score)
)
df.loc[df['PSQI'], 'latency_score2'] = df['mtc_12a1'] - 1

latency_sum = df['latency_score1'] + df['latency_score2']
df.loc[df['PSQI'], 'C2'] = pd.cut(
    latency_sum, [-1, 0, 2, 4, 6], labels=[0, 1, 2, 3]
).astype(float)

# C3. 수면 시간
def sleep_duration_score(x):
    if x >= 420: return 0
    elif x >= 360: return 1
    elif x >= 300: return 2
    else: return 3

df.loc[df['PSQI'], 'C3'] = (
    df.loc[df['PSQI'], 'sleep_time_min']
    .apply(sleep_duration_score)
)

# C4. 수면 효율
df.loc[df['PSQI'], 'sleep_efficiency'] = (
    df['sleep_time_min'] / df['time_in_bed_min'] * 100
)

def efficiency_score(x):
    if x >= 85: return 0
    elif x >= 75: return 1
    elif x >= 65: return 2
    else: return 3

df.loc[df['PSQI'], 'C4'] = (
    df.loc[df['PSQI'], 'sleep_efficiency']
    .apply(efficiency_score)
)

# C5. 수면 방해
disturbance_variables = [
    'mtc_12b1','mtc_12c1','mtc_12d1','mtc_12e1',
    'mtc_12f1','mtc_12g1','mtc_12h1','mtc_12i1','mtc_12j1'
]

df.loc[df['PSQI'], 'disturbance_sum'] = (
    df.loc[df['PSQI'], disturbance_variables]
    .apply(lambda x: x - 1)
    .sum(axis=1)
)

df.loc[df['PSQI'], 'C5'] = pd.cut(
    df['disturbance_sum'], [-1, 0, 9, 18, 27], labels=[0, 1, 2, 3]
).astype(float)

# C6. 수면제 사용
df.loc[df['PSQI'], 'C6'] = df['mtc_14z1'] - 1

# C7. 주간 기능 장애
daytime_sum = (df['mtc_15z1'] - 1) + (df['mtc_16z1'] - 1)
df.loc[df['PSQI'], 'C7'] = pd.cut(
    daytime_sum, [-1, 0, 2, 4, 6], labels=[0, 1, 2, 3]
).astype(float)

# ======================================================
# 4. PSQI-K 총점 및 해석
# ======================================================
components = ['C1','C2','C3','C4','C5','C6','C7']

df.loc[df['PSQI'], 'PSQI_KR_Score'] = df[components].sum(axis=1)

def psqi_status(x):
    if pd.isna(x): return np.nan
    elif x >= 10: return '불면증/수면장애 가능성 높음'
    elif x >= 5:  return '수면의 질 저하 가능성'
    else:         return '정상'

df.loc[df['PSQI'], 'PSQI_KR_Status'] = (
    df.loc[df['PSQI'], 'PSQI_KR_Score']
    .apply(psqi_status)
)

# ======================================================
# 5. 유효하지 않은 응답자는 전부 NaN 처리
# ======================================================
df.loc[~df['PSQI'],
       components + ['PSQI_KR_Score', 'PSQI_KR_Status']] = np.nan




#PS_QI_KR_Status를 숫자 012 로----------------------------------------------------------------------------------------------------------------
conditions = [
    df['PSQI_KR_Status']  == '정상',
    df['PSQI_KR_Status'] == '수면의 질 저하 가능성',
    df['PSQI_KR_Status']  == '불면증/수면장애 가능성 높음',
    
]

choices = [
    0,          
    1,          
    2   
]

df['PSQI_KR_Status_cat'] = np.select(conditions, choices, default=np.nan)
```

```python
#흡연 여부-----------------------------------------------------------------------------------------------------------------
# 하루 평균 흡연량(연속형)
df['smoke_amount'] = df['cig_per_day'].replace([777, 888, 999], np.nan)

m = df['smoke_amount'].notna()
q99 = df.loc[m, 'smoke_amount'].quantile(0.99)
df.loc[m, 'smoke_amount'] = df.loc[m, 'smoke_amount'].clip(upper=q99)

# 흡연 범주형
df['smoke_cat'] = pd.cut(
    df['smoke_amount'],
    bins=[0, 1, 10, 20, 30, np.inf],
    labels=['0','1–9', '10–19', '20–29', '≥30'],
    right=False
)



# 흡연자 여부
df['smoker'] = np.nan
df['smoking_lifetime'] = df['smoking_lifetime'].replace({7: np.nan, 9: np.nan})
df['smoking_current'] = df['smoking_current'].replace({7: np.nan, 9: np.nan})


# (1) sma_01z1=2 AND sma_03z2 in (1,2)  -> 흡연자
df.loc[(df['smoking_lifetime'] == 2) & (df['smoking_current'].isin([1, 2])), 'smoker'] = 1

# (2) sma_01z1=2 AND sma_03z2=3 -> 비흡연자
df.loc[(df['smoking_lifetime'] == 2) & (df['smoking_current'] == 3), 'smoker'] = 0

# (3) sma_01z1 in (1,3,8) -> 비흡연자 (sma_03z2 값과 무관)
df.loc[df['smoking_lifetime'].isin([1, 3, 8]), 'smoker'] = 0

df.loc[df['smoker'] == 0, ['smoke_amount', 'smoke_cat']] = np.nan
```

#### 3. 건강상태/기능(만성질환/BMI/우울상태/미충족의료)

## 기술통계

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

WEIGHT = "wt_p"
PSU = "SPOT_NO"
def survey_ols(formula, data, weight=WEIGHT, psu=PSU):
    y, X = patsy.dmatrices(formula, data=data.replace([np.inf,-np.inf], np.nan),
                           return_type="dataframe")

    d = data.loc[y.index, [weight, psu]].copy()
    d = d.dropna(subset=[weight, psu])
    d = d[d[weight] > 0]

    y = y.loc[d.index]
    X = X.loc[d.index]
    
    model = sm.WLS(y, X, weights=d[weight]) 
    res = model.fit(cov_type="cluster", cov_kwds={"groups": d[psu].astype(str)})
    return res


def pval_cat_weighted_chi2(df, x, group_col, weight_col):
    
    d = df[[x, group_col, weight_col]].dropna()
    if d[group_col].nunique() < 2 or d[x].nunique() < 2:
        return np.nan


    tab = (d.pivot_table(index=x, columns=group_col, values=weight_col,
                         aggfunc="sum", fill_value=0.0))

  
    chi2, p, dof, exp = chi2_contingency(tab.to_numpy(), correction=False)
    return float(p)

def w_quantile(x, w, qs):
    x = np.asarray(x)
    w = np.asarray(w)
    m = np.isfinite(x) & np.isfinite(w)
    if m.sum() == 0:
        return [np.nan]*len(qs)
    x = x[m]; w = w[m]
    order = np.argsort(x)
    x = x[order]; w = w[order]
    cw = np.cumsum(w)
    cw = cw / cw[-1]
    return [x[np.searchsorted(cw, q, side="left")] for q in qs]

def cont_stats(s, w, na_codes=None):
    s = pd.to_numeric(s, errors="coerce")
    w = pd.to_numeric(w, errors="coerce")

    if na_codes is not None:
        s = s.replace(na_codes, np.nan)

    m = s.notna() & w.notna()

    unwt_n = int(m.sum())
    nmiss  = int((~m).sum())   
    
    if unwt_n == 0:
        return dict(Unweighted_N=0, Nmiss=nmiss, Weighted_N=np.nan,
                    Weighted_Mean=np.nan, SE=np.nan,
                    Weighted_Median=np.nan, IQR=np.nan)

    x = s[m].to_numpy(float)
    ww = w[m].to_numpy(float)
    sumw = ww.sum()
    wmean = np.average(x, weights=ww)

    wvar = np.average((x - wmean)**2, weights=ww)
    neff = (sumw**2) / (np.sum(ww**2))
    se = np.sqrt(wvar / neff)

    q25, q50, q75 = w_quantile(x, ww, [0.25, 0.5, 0.75])
    iqr = q75 - q25

    return dict(Unweighted_N=unwt_n, Nmiss=nmiss, Weighted_N=sumw,
                Weighted_Mean=wmean, SE=se,
                Weighted_Median=q50, IQR=iqr)

def cat_rows(s, w, dropna=True, na_codes=None):
    s = s.copy()
    w = pd.to_numeric(w, errors="coerce")

    if na_codes is not None:
        s = s.replace(na_codes, np.nan)

    if dropna:
        m = s.notna() & w.notna()
        s = s[m]
        w = w[m]

    vc_u = s.value_counts(dropna=False)

    dfw = pd.DataFrame({"s": s, "w": w}).dropna(subset=["w"])
    wsum_total = dfw["w"].sum()
    wsum_by = dfw.groupby("s")["w"].sum()
    wpct_by = (wsum_by / wsum_total) * 100 if wsum_total > 0 else wsum_by*np.nan

    cats = list(vc_u.index)
    out = []
    for c in cats:
        u_n = int(vc_u.get(c, 0))
        w_p = float(wpct_by.get(c, np.nan))
        out.append((c, u_n, w_p))
    return out

def fmt_n_pct(n, pct, digits=1):
    if pd.isna(n) or pd.isna(pct):
        return ""
    
    return f"{int(round(n)):,} ({pct:.{digits}f}%)"

def fmt_num(x, digits=3):
    return "" if pd.isna(x) else f"{x:.{digits}f}"



def fmt_p(p):
    if p is None or (isinstance(p, float) and np.isnan(p)):
        return ""
    if p < 0.001:
        return "<0.001"
    return f"{p:.3f}"

def pval_cont_survey_ols(df, var, group_col, weight_col, survey_ols):
    d = df[[var, group_col, weight_col, PSU]].copy()
    d = d.dropna(subset=[var, group_col, weight_col, PSU])
    if d[group_col].nunique() < 2:
        return np.nan

    fit = survey_ols(f"{var} ~ C({group_col})", d)

    pv = getattr(fit, "pvalues", None)
    if pv is None:
        return np.nan

    candidates = [
        f"C({group_col})[T.1]",
        f"C({group_col})[T.1.0]",
        f"C({group_col})[T.True]",
    ]
    for k in candidates:
        if k in pv.index:
            return float(pv[k])

 
    idx = [i for i in pv.index if i.startswith(f"C({group_col})[T.")]
    return float(pv[idx[0]]) if idx else np.nan

# --------------------------
# 메인: Table  생성
# --------------------------
def build_qc_table1(df, group_col, weight_col,
                    cont_vars, cat_vars,
                    eq5d_hi_col=None, eq5d_lo_col=None,
                    survey_ols=None,
                    pval_cat_func=None):
    
    base = df[df[group_col].isin([0, 1])].copy()
    d_all = base
    d_nor = base[base[group_col] == 0]
    d_dep = base[base[group_col] == 1]

    rows = []
    sub = "\u00A0\u00A0"

    def add_header(var_label, pval=""):
        rows.append({"Variable": var_label, "All":"", "Normal":"", "Depression":"", "p-value": pval})

    def add_subrow(stat_label, all_v, nor_v, dep_v):
        rows.append({
            "Variable": f"{sub}{stat_label}",
            "All": all_v, "Normal": nor_v, "Depression": dep_v,
            "p-value": ""
        })

    cont_vars = cont_vars or []
    cat_vars = cat_vars or []


    if eq5d_hi_col is not None and eq5d_lo_col is not None:
        add_header("EQ-5D", pval="")

        def one_rate_cell(d, col):
            rr = cat_rows(d[col], d[weight_col])  
            for c, u_n, w_p in rr:
                if c == 1 or c == 1.0 or str(c) == "1":
                    return f"{u_n:,} ({w_p:.1f}%)"
            return ""

        add_subrow("high", one_rate_cell(d_all, eq5d_hi_col),
                         one_rate_cell(d_nor, eq5d_hi_col),
                         one_rate_cell(d_dep, eq5d_hi_col))
        add_subrow("low",  one_rate_cell(d_all, eq5d_lo_col),
                         one_rate_cell(d_nor, eq5d_lo_col),
                         one_rate_cell(d_dep, eq5d_lo_col))

  
    for v, label in cont_vars:
        p_str = ""
        if survey_ols is not None:
            p = pval_cont_survey_ols(d_all, v, group_col, weight_col, survey_ols)
            p_str = fmt_p(p)

        add_header(label, pval=p_str)

        na_cont = NA_CONT_MAP.get(v, None)

        s_all = cont_stats(d_all[v], d_all[weight_col], na_codes=na_cont)
        s_nor = cont_stats(d_nor[v], d_nor[weight_col], na_codes=na_cont)
        s_dep = cont_stats(d_dep[v], d_dep[weight_col], na_codes=na_cont)

        add_subrow("Unweighted N (Nmiss)",
                   f"{s_all['Unweighted_N']:,} ({s_all['Nmiss']:,})",
                   f"{s_nor['Unweighted_N']:,} ({s_nor['Nmiss']:,})",
                   f"{s_dep['Unweighted_N']:,} ({s_dep['Nmiss']:,})")

        add_subrow("Weighted N (sum w)",
                   fmt_num(s_all["Weighted_N"], 3),
                   fmt_num(s_nor["Weighted_N"], 3),
                   fmt_num(s_dep["Weighted_N"], 3))

        add_subrow("Weighted Mean ± SE",
                   f"{fmt_num(s_all['Weighted_Mean'],3)} ± {fmt_num(s_all['SE'],3)}",
                   f"{fmt_num(s_nor['Weighted_Mean'],3)} ± {fmt_num(s_nor['SE'],3)}",
                   f"{fmt_num(s_dep['Weighted_Mean'],3)} ± {fmt_num(s_dep['SE'],3)}")

        add_subrow("Weighted Median (IQR)",
                   f"{fmt_num(s_all['Weighted_Median'],3)} ({fmt_num(s_all['IQR'],3)})",
                   f"{fmt_num(s_nor['Weighted_Median'],3)} ({fmt_num(s_nor['IQR'],3)})",
                   f"{fmt_num(s_dep['Weighted_Median'],3)} ({fmt_num(s_dep['IQR'],3)})")

 
    for v, label in cat_vars:
    # p-값
        p_str = ""
        if pval_cat_func is not None:
            try:
                p = pval_cat_func(d_all, v, group_col, weight_col)
                p_str = fmt_p(p)
            except Exception:
                p_str = ""

        na_cat = NA_CAT_MAP.get(v, None)

      
        if na_cat is None:
            N_total = int(d_all[v].notna().sum())
        else:
            N_total = int(d_all[v].replace(na_cat, np.nan).notna().sum())

      
        add_header(f"{label} (N={N_total:,})", pval=p_str)

        all_rows = cat_rows(d_all[v], d_all[weight_col], na_codes=na_cat)
        nor_rows = cat_rows(d_nor[v], d_nor[weight_col], na_codes=na_cat)
        dep_rows = cat_rows(d_dep[v], d_dep[weight_col], na_codes=na_cat)

        cats = [r[0] for r in all_rows]

        def to_dict(rr):
            return {c: (u_n, w_p) for (c, u_n, w_p) in rr}

        A = to_dict(all_rows); N = to_dict(nor_rows); D = to_dict(dep_rows)

        for c in cats:
            a_u, a_wp = A.get(c, (np.nan, np.nan))
            n_u, n_wp = N.get(c, (np.nan, np.nan))
            d_u, d_wp = D.get(c, (np.nan, np.nan))

            all_cell = "" if pd.isna(a_u) else f"{int(a_u):,} ({a_wp:.1f}%)"
            nor_cell = "" if pd.isna(n_u) else f"{int(n_u):,} ({n_wp:.1f}%)"
            dep_cell = "" if pd.isna(d_u) else f"{int(d_u):,} ({d_wp:.1f}%)"

            add_subrow(str(c), all_cell, nor_cell, dep_cell)
        out = pd.DataFrame(rows, columns=["Variable","All","Normal","Depression","p-value"])
    return out
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import re

WCOL = "wt_p"
CLUSTER = "SPOT_NO"
Y = "eq5d_mean_bin"
G= "PHQ9_2cat" 

def fit_glm_logit_cluster(formula, data, wcol=WCOL, cluster_col=CLUSTER):
    d = data.copy()
    d = d.dropna(subset=[wcol, cluster_col])
    res = smf.glm(
        formula=formula,
        data=d,
        family=sm.families.Binomial(),
        freq_weights=d[wcol]
    ).fit(
        cov_type="cluster",
        cov_kwds={"groups": d[cluster_col]}
    )
    return res

def extract_or_table(res, x_term):
    t = pd.DataFrame({
        "term": res.params.index,
        "coef": res.params.values,
        "SE": res.bse.values,
        "p": res.pvalues.values
    })
    t["OR"] = np.exp(t["coef"])
    t["CI95_low"]  = np.exp(t["coef"] - 1.96 * t["SE"])
    t["CI95_high"] = np.exp(t["coef"] + 1.96 * t["SE"])
    t = t[t["term"] != "Intercept"].copy()
    t = t[t["term"].str.contains(x_term, regex=False)].copy()
    return t

def _base(term):
    return term.replace("C(", "").replace(")", "").strip()

def _overall_p_for_x(res, x, x_base):
    
    if "C(" not in x:   
        return None
    try:
        w = res.wald_test_terms(skip_single=False)
     
        idx = list(w.table.index.astype(str))
        target1 = f"C({x_base})"
        target2 = x_base
        if target1 in idx:
            return float(w.table.loc[target1, "pvalue"])
        if target2 in idx:
            return float(w.table.loc[target2, "pvalue"])
    
        for k in idx:
            if k == target1 or k == target2 or k.endswith(f"({x_base})"):
                return float(w.table.loc[k, "pvalue"])
    except Exception:
        pass
    return None

def logit_all_vars_adj(df_2, x_list, covars, add_overall_p=True):
    
    out_rows = []
    covars = list(covars)

    for x in x_list:
        x_base = _base(x)

        rhs_terms = [x] + covars
        rhs_terms = list(dict.fromkeys(rhs_terms)) 
        formula = f"{Y} ~ " + " + ".join(rhs_terms)

        
        need_cols = [Y, WCOL, CLUSTER, x_base]
        for v in covars:
            need_cols.append(_base(v))
        need_cols = list(dict.fromkeys(need_cols))

        dfit = df_2.dropna(subset=need_cols).copy()

        if dfit.shape[0] == 0:
            out_rows.append({"model":"adj", "x":x, "term":np.nan, "OR":np.nan,
                             "CI95_low":np.nan, "CI95_high":np.nan, "p":np.nan,
                             "overall_p":np.nan, "N":0, "note":"no data", "formula":formula})
            continue
        if dfit[Y].nunique() < 2:
            out_rows.append({"model":"adj", "x":x, "term":np.nan, "OR":np.nan,
                             "CI95_low":np.nan, "CI95_high":np.nan, "p":np.nan,
                             "overall_p":np.nan, "N":int(dfit.shape[0]), "note":"Y has <2 levels",
                             "formula":formula})
            continue

        try:
            res = fit_glm_logit_cluster(formula, dfit)
            tab = extract_or_table(res, x_term=x_base)
            if tab.shape[0] == 0:
                tab = extract_or_table(res, x_term=x) 

            overall_p = _overall_p_for_x(res, x, x_base) if add_overall_p else None

            for _, r in tab.iterrows():
                out_rows.append({
                    "model": "adj",
                    "x": x,
                    "term": r["term"],
                    "OR": r["OR"],
                    "CI95_low": r["CI95_low"],
                    "CI95_high": r["CI95_high"],
                    "p": r["p"],
                    "overall_p": overall_p,
                    "N": int(res.nobs),
                    "note": "",
                    "formula": formula
                })

        except Exception as e:
            out_rows.append({"model":"adj", "x":x, "term":np.nan, "OR":np.nan,
                             "CI95_low":np.nan, "CI95_high":np.nan, "p":np.nan,
                             "overall_p":np.nan, "N":int(dfit.shape[0]),
                             "note": f"fail: {type(e).__name__}", "formula":formula})

    return pd.DataFrame(out_rows)

def sort_terms_within_x(df_out, x_order=None):
    d = df_out.copy()
    if x_order is not None:
        d["x"] = pd.Categorical(d["x"], categories=x_order, ordered=True)

    def extract_level(term):
        m = re.search(r"\[T\.(.+?)\]$", str(term))
        return m.group(1) if m else ""

    d["level"] = d["term"].apply(extract_level)
    def level_key(v):
        s = str(v)
        try:
            return (0, float(s))
        except:
            return (1, s)
    d["level_key"] = d["level"].apply(level_key)
    d["is_dummy"] = d["term"].astype(str).str.contains(r"\[T\.", regex=True).astype(int)

    d = d.sort_values(["x", "is_dummy", "level_key", "term"]).reset_index(drop=True)
    return d.drop(columns=["level","level_key","is_dummy"], errors="ignore")
```

```python

WCOL = "wt_p"
CLUSTER = "SPOT_NO"   

Y = "eq5d_mean_bin"        
G = "PHQ9_2cat"       



def fit_glm_logit_cluster(formula, data, wcol=WCOL, cluster_col=CLUSTER):
    """
    GLM Binomial 로지스틱 + freq_weights(가중치) + cluster-robust SE(집락)
    """
    d = data.copy()

    
    d = d.dropna(subset=[wcol, cluster_col])

    res = smf.glm(
        formula=formula,
        data=d,
        family=sm.families.Binomial(),
        freq_weights=d[wcol]
    ).fit(
        cov_type="cluster",
        cov_kwds={"groups": d[cluster_col]}
    )
    return res


def extract_or_table(res, x_term):
  
    params = res.params
    bse = res.bse
    pvals = res.pvalues

    t = pd.DataFrame({"term": params.index, "coef": params.values})
    t["SE"] = [bse[k] for k in t["term"]]
    t["p"] = [pvals[k] for k in t["term"]]
    t["OR"] = np.exp(t["coef"])
    t["CI95_low"] = np.exp(t["coef"] - 1.96*t["SE"])
    t["CI95_high"] = np.exp(t["coef"] + 1.96*t["SE"])


    t = t[t["term"] != "Intercept"].copy()
    t = t[t["term"].str.contains(x_term, regex=False)].copy()

    return t


def univariate_logit_table(df_2, x_list, group_value=None, group_col=G):
   
    if group_value is not None:
        dsub = df_2[df_2[group_col] == group_value].copy()
        group_name = "Normal" if group_value == 0 else "Depression"
    else:
        dsub = df_2.copy()
        group_name = "All"

    out_rows = []

    for x in x_list:
        formula = f"{Y} ~ {x}"

        
        cols_need = [Y, WCOL, CLUSTER]
        x_base = x.replace("C(", "").replace(")", "").strip()
        cols_need.append(x_base)

        dfit = dsub.dropna(subset=list(dict.fromkeys(cols_need))).copy()

        
        if dfit.shape[0] == 0:
            out_rows.append({"group": group_name, "x": x, "term": np.nan, "OR": np.nan,
                             "CI95_low": np.nan, "CI95_high": np.nan, "p": np.nan,
                             "N": 0, "note": "no data"})
            continue
        if dfit[Y].nunique() < 2:
            out_rows.append({"group": group_name, "x": x, "term": np.nan, "OR": np.nan,
                             "CI95_low": np.nan, "CI95_high": np.nan, "p": np.nan,
                             "N": dfit.shape[0], "note": "Y has <2 levels"})
            continue

        try:
            res = fit_glm_logit_cluster(formula, dfit)
            tab = extract_or_table(res, x_term=x_base)
            if tab.shape[0] == 0:
                tab = extract_or_table(res, x_term=x)

            for _, r in tab.iterrows():
                out_rows.append({
                    "group": group_name,
                    "x": x,
                    "term": r["term"],
                    "OR": r["OR"],
                    "CI95_low": r["CI95_low"],
                    "CI95_high": r["CI95_high"],
                    "p": r["p"],
                    "N": int(res.nobs),
                    "note": ""
                })

        except Exception as e:
            out_rows.append({
                "group": group_name,
                "x": x,
                "term": np.nan,
                "OR": np.nan,
                "CI95_low": np.nan,
                "CI95_high": np.nan,
                "p": np.nan,
                "N": int(dfit.shape[0]),
                "note": f"fail: {type(e).__name__}"
            })

    out = pd.DataFrame(out_rows)
    return out
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
def forestplot_two_groups(
    df,
    var_order=None,
    group_order=("Normal", "Depression"),
    group_palette=("tab:blue", "tab:red"),
    title="Forest plot (Adjusted OR with 95% CI)",
    x_label="Odds Ratio",
    y_label="",
    figsize=(8, 4.8),
    xlim=None,
    add_value_text=True
):
    """
    df columns required:
      - Variable
      - group
      - OR
      - CI95_low
      - CI95_high
    """
    dfp = df.copy()
    # (1) 필수 컬럼 체크
    required = {"Variable", "group", "OR", "CI95_low", "CI95_high"}
    missing = required - set(dfp.columns)
    if missing:
        raise ValueError(f"Missing columns: {missing}")
    # (2) OR/CI 숫자형 변환
    for c in ["OR", "CI95_low", "CI95_high"]:
        dfp[c] = pd.to_numeric(dfp[c], errors="coerce")
    # (3) 변수 순서 고정
    if var_order is None:
        # 기본: 데이터에 등장하는 순서대로
        var_order = list(dfp["Variable"].dropna().unique())
    dfp["Variable"] = pd.Categorical(dfp["Variable"], categories=var_order, ordered=True)
    # (4) group 순서 고정
    dfp["group"] = pd.Categorical(dfp["group"], categories=list(group_order), ordered=True)
    # (5) y축 위치(변수별)
    # seaborn이 categorical y를 처리하지만, 에러바를 깔끔하게 겹치려면 y좌표를 수동 생성
    y_map = {v: i for i, v in enumerate(var_order)}
    dfp["y_base"] = dfp["Variable"].map(y_map).astype(float)
    # 그룹별로 y를 조금씩 위/아래로 오프셋(겹침 방지)
    offsets = np.linspace(-0.15, 0.15, num=len(group_order))
    offset_map = {g: offsets[i] for i, g in enumerate(group_order)}
    dfp["y"] = dfp["y_base"] + dfp["group"].map(offset_map).astype(float)
    # (6) 플롯 시작
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=figsize)
    # 기준선 OR=1
    ax.axvline(1.0, linestyle="--", linewidth=1)
    # (7) 그룹별 점 + 에러바
    for g, color in zip(group_order, group_palette):
        sub = dfp[dfp["group"] == g].dropna(subset=["OR", "CI95_low", "CI95_high"])
        # 에러바(가로)
        x = sub["OR"].values
        y = sub["y"].values
        xerr_low = x - sub["CI95_low"].values
        xerr_high = sub["CI95_high"].values - x
        xerr = np.vstack([xerr_low, xerr_high])
        ax.errorbar(
            x, y,
            xerr=xerr,
            fmt="o",
            color=color,
            ecolor=color,
            elinewidth=1.5,
            capsize=3,
            markersize=6,
            label=str(g)
        )
        if add_value_text:
            # OR(95%CI) 텍스트를 오른쪽에 가볍게 붙임
            for xi, yi, lo, hi in zip(x, y, sub["CI95_low"].values, sub["CI95_high"].values):
                ax.text(
                    xi, yi,
                    f" {xi:.3f} ({lo:.3f}-{hi:.3f})",
                    va="center",
                    ha="left",
                    fontsize=8,
                    color=color
                )
    # (8) y축 라벨(변수명)
    ax.set_yticks(range(len(var_order)))
    ax.set_yticklabels(var_order)
    # (9) 로그 스케일 권장(OR 플롯 표준)
    if xlim:
        ax.set_xlim(xlim)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_title(title)
    ax.legend(title="Group", loc="best")
    plt.tight_layout()
    plt.show()
    return fig, ax
# df_result
label_map = {
    "Physical activity (sufficient vs insufficient)": "Sufficient physical activity "
}
table3_sorted["Variable"] =table3_sorted["Variable"].replace(label_map)
10:42
var_order = [
    #"Sufficient physical activity (vs insufficient)",
    #"Tooth brushing after dinner",
    #"Binge drinking",
    #"Current smoker",
    #"Diet control",
    "Abnormal BMI",
    "Chronic diseases",
    "PSQI score (per 1-point increase)",
]
10:43
forestplot_two_groups(
    df=table3_sorted,
    var_order=var_order[::-1],
    group_order=("Normal", "Depression"),
    title="Health behaviors/status and EQ-5D (>= mean = 1): Adjusted OR",
    xlim=(0.6, 2.5),   # 필요 시 조정
    add_value_text=False
)
10:44
#히트맵
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
def make_group_heatmap(
    table3_sorted: pd.DataFrame,
    var_order: list,
    group_order=("Normal", "Depression"),
    alpha=0.05,
    title="Association direction with EQ-5D (>= mean = 1)",
    figsize=(5.2, 4.8),
):
    """
    table3_sorted columns required:
      - Variable, group, OR, p
    Logic:
      - p >= alpha  -> 0 (NS, gray)
      - p < alpha & OR > 1 -> +1 (positive association, blue)
      - p < alpha & OR < 1 -> -1 (negative association, red)
    """
    df = table3_sorted.copy()
    # 1) p-value 숫자형 변환 (문자 '<0.001' 같은 게 있으면 처리)
    def parse_p(x):
        if pd.isna(x):
            return np.nan
        if isinstance(x, str):
            x = x.strip()
            if x.lower() in ["ns", "n.s.", "not significant"]:
                return 1.0
            if x.startswith("<"):
                # 예: "<0.001"
                try:
                    return float(x.replace("<", ""))  # 0.001
                except:
                    return np.nan
            try:
                return float(x)
            except:
                return np.nan
        return float(x)
    df["p_num"] = df["p"].apply(parse_p)
    # 2) 방향성 코드 생성: -1 / 0 / +1
    df["dir"] = 0
    sig = df["p_num"] < alpha
    df.loc[sig & (df["OR"] > 1), "dir"] = 1
    df.loc[sig & (df["OR"] < 1), "dir"] = -1
    # 3) 히트맵용 pivot (rows=Variable, cols=group)
    pivot = (
        df.pivot_table(index="Variable", columns="group", values="dir", aggfunc="first")
        .reindex(index=var_order)
        .reindex(columns=list(group_order))
    )
    # 4) 표시용 라벨(옵션): OR 값 혹은 NS
    ann = (
        df.assign(label=np.where(df["p_num"] < alpha,
                                 df["OR"].map(lambda v: f"{v:.2f}"),
                                 "NS"))
          .pivot_table(index="Variable", columns="group", values="label", aggfunc="first")
          .reindex(index=var_order)
          .reindex(columns=list(group_order))
    )
    # 5) 컬러맵: -1=red, 0=gray, +1=blue
    # seaborn은 연속값 컬러맵을 쓰므로, center=0으로 두고 회색이 중간에 오도록 설정
    cmap = sns.diverging_palette(10, 240, as_cmap=True)
    plt.figure(figsize=figsize)
    ax = sns.heatmap(
        pivot,
        cmap=cmap,
        center=0,
        vmin=-1, vmax=1,
        linewidths=0.7,
        linecolor="white",
        cbar=False,
        annot=ann,
        fmt="",
        annot_kws={"fontsize": 10}
    )
    ax.set_title(title, fontsize=12, pad=10)
    ax.set_xlabel("")
    ax.set_ylabel("")
    # 축 라벨 정리
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)
    # 범례 설명
    plt.figtext(
        0.02, 0.01,
        f"Blue: OR>1 & p<{alpha} (positive association)\n"
        f"Red: OR<1 & p<{alpha} (negative association)\n"
        f"Gray: NS",
        ha="left", fontsize=9
    )
    plt.tight_layout()
    plt.show()
    return pivot, ann
# var_order는 PPT에서 보여줄 순서대로 넣으면 됨
var_order = [
    "Sufficient physical activity (vs insufficient)",
    "Tooth brushing after dinner",
    "Binge drinking",
    "Current smoker",
    "Diet control",
    "Abnormal BMI",
    "Chronic diseases",
    "PSQI score (per 1-point increase)",
]
# table3_sorted
pivot_dir, pivot_label = make_group_heatmap(
    table3_sorted,
    var_order=var_order,
    group_order=("Normal", "Depression"),
    alpha=0.05,
    title="Health behaviors/status vs EQ-5D (>= mean = 1): Direction & Significance"
)
```
